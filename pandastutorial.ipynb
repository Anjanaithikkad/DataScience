```python
import pandas as pd

df = pd.read_csv('sample.csv')

for x in df.index:
    if df.loc[x, "Duration"] > 120:
        df.loc[x, "Duration"] = 120

print(df.to_string())
```

```python
import pandas as pd

df = pd.read_csv('sample.csv')

for x in df.index:
      if df.loc[x, "Duration"] > 120:
        df.drop(x, inplace = True)
print(df.to_string())
```
